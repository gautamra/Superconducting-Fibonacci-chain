## u_0 vs V phase diagram

```python
comp_u0V = pickle.load(open("comp_u0V.p","rb"))
comp_u0VChris = pickle.load(open("comp_u0VChris.p","rb"))
comp_u0Vold = pickle.load(open("comp_u0Vold.p","rb"))
comp_u0V = {**comp_u0V, **comp_u0VChris, **comp_u0Vold}
keys = sorted(comp_u0V.keys())
[print(key) for key in keys]
```

```python
Vs = []
u0_inds = []
for key in keys:
    V,u0_ind = key
    if V not in Vs:
        Vs.append(V)
    if u0_ind not in u0_inds:
        u0_inds.append(u0_ind)
u0s = [u0_ind/100 for u0_ind in u0_inds]

# Vs = Vs[1:]
```

```python
DeltaAvgArray = []
for u0_ind in u0_inds:
    DeltaAvgList = []
    for V in Vs:
        DeltaAvg = np.average(comp_u0V[(V,u0_ind)]["Delta"])
        DeltaAvgList.append(DeltaAvg.real)
    DeltaAvgArray.append(DeltaAvgList)
```

```python
fig, ax = plt.subplots()

ax.contourf(Vs, u0s ,DeltaAvgArray)
```

```python
fig, ax = plt.subplots()

ax.imshow(DeltaAvgArray, origin = 'lower', extent = (min(Vs),max(Vs),min(u0s),max(u0s)))
```

```python
Vs = np.array(Vs)
Vlines = (Vs[:-1]+Vs[1:])/2
Vlines = np.hstack((Vlines[0]-(Vlines[1]-Vlines[0]),Vlines, 2*Vlines[len(Vlines)-1] - Vlines[len(Vlines)-2]))

fig, axes = plt.subplots(1,2,gridspec_kw={'width_ratios':[16,1]})
ax, axcb = axes

cmap = mpl.cm.get_cmap('viridis')
Deltamax = max(max(DeltaAvgArray))

ax.set_xlim((Vlines[0],Vlines[-1]))
ax.set_ylim((0,1.16))
ax.set_xlabel("$V$")
ax.set_ylabel("$u_0$")

for iv in range(len(Vs[:])):
    u0s = []
    for key in keys:
        (V,u0_ind) = key
        if V == Vs[iv]:
            u0s.append(u0_ind/100)
    u0s = np.array(u0s)
    u0lines = (u0s[:-1]+u0s[1:])/2
    u0lines = np.hstack((u0lines[0]-(u0lines[1]-u0lines[0]),u0lines, 2*u0lines[len(u0lines)-1] - u0lines[len(u0lines)-2]))
    for iu0 in range(len(u0s)):

        color = cmap(DeltaAvgArray[iu0][iv]/Deltamax)
        rect = plt.Rectangle((Vlines[iv], u0lines[iu0]), Vlines[iv+1]-Vlines[iv], u0lines[iu0+1]-u0lines[iu0],
                     facecolor=color)
        ax.add_patch(rect) 

# fig, ax = plt.subplots(figsize=(6, 1))
# fig.subplots_adjust(bottom=0.5)


norm = mpl.colors.Normalize(vmin=0, vmax=Deltamax)
cb1 = mpl.colorbar.ColorbarBase(axcb, cmap=cmap,
                                norm=norm,
                                orientation='vertical')
cb1.set_label('$\\Delta_{avg}$')
```
